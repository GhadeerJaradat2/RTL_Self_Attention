```python
##convert content to hex

def signed_int_to_8bit_hex(value):
    """Convert a signed integer to an 8-bit hexadecimal string."""
    if value < 0:
        value = (1 << 8) + value  # Two's complement for negative numbers
    return format(value & 0xFF, '02x')  # Format as 8-bit hexadecimal

def convert_file_to_hex(input_file_path, output_file_path):
    with open(input_file_path, 'r') as infile, open(output_file_path, 'w') as outfile:
        for line in infile:
            # Split the line into integers
            integers = [int(num) for num in line.split()]

            # Convert each integer to 8-bit hex
            hex_values = [signed_int_to_8bit_hex(num) for num in integers]

            # Write the hex values to the output file, maintaining line structure
            outfile.write(' '.join(hex_values) + '\n')



# Example usage
          
input_file_path = 'C:/Users/100062811/OneDrive - ku.ac.ae/Ghadeer/MatrixMulFullExample/INTXINT_Result.txt'  # Replace with your input file path
output_file_path = 'C:/Users/100062811/OneDrive - ku.ac.ae/Ghadeer/MatrixMulFullExample/INTXINT_Result_hex.txt'  # Replace with your output file path
convert_file_to_hex(input_file_path, output_file_path)
print('Conversion complete.')
```

```python
def float_to_fixed_point_hex(value, fraction_bits=8):
    """Convert a floating-point number to a fixed-point hexadecimal string."""
    # Convert the float to a fixed-point representation
    fixed_point_value = int(value * (1 << fraction_bits))

    # Handle two's complement for negative numbers
    if fixed_point_value < 0:
        fixed_point_value = (1 << (8 + fraction_bits)) + fixed_point_value

    # Format as hexadecimal
    hex_str = format(fixed_point_value & ((1 << (8 + fraction_bits)) - 1), '04x')
    return hex_str

def convert_file_to_hex_with_fraction(input_file_path, output_file_path, fraction_bits=8):
    with open(input_file_path, 'r') as infile, open(output_file_path, 'w') as outfile:
        for line in infile:
            # Split the line into floats
            floats = [float(num) for num in line.split()]

            # Convert each float to fixed-point hex
            hex_values = [float_to_fixed_point_hex(num, fraction_bits) for num in floats]

            # Write the hex values to the output file, maintaining line structure
            outfile.write(' '.join(hex_values) + '\n')

# Example usage
input_file_path = 'C:/Users/100062811/OneDrive - ku.ac.ae/Ghadeer/MatrixMulFullExample/INTXFrac_Result.txt'  
output_file_path ='C:/Users/100062811/OneDrive - ku.ac.ae/Ghadeer/MatrixMulFullExample/INTXFrac_Result_hex.txt'
convert_file_to_hex_with_fraction(input_file_path, output_file_path)
print('Conversion complete.')
```

```python
def fractional_to_fixed_point_hex(value, total_bits=8):
    """Convert a purely fractional number to a fixed-point hexadecimal string."""
   

    # Convert the fraction to a fixed-point representation
    fixed_point_value = int(value * (1 << total_bits))

    # Format as hexadecimal
    hex_str = format(fixed_point_value & ((1 << total_bits) - 1), '02x')
    return hex_str

def convert_file_to_hex_pure_fraction(input_file_path, output_file_path, total_bits=8):
    with open(input_file_path, 'r') as infile, open(output_file_path, 'w') as outfile:
        for line in infile:
            # Split the line into fractional numbers
            fractions = [float(num) for num in line.split()]

            # Convert each fraction to fixed-point hex
            hex_values = [fractional_to_fixed_point_hex(num, total_bits) for num in fractions]

            # Write the hex values to the output file, maintaining line structure
            outfile.write(' '.join(hex_values) + '\n')

# Example usage
input_file_path = 'C:/Users/100062811/OneDrive - ku.ac.ae/Ghadeer/MatrixMulFullExample/Query_Fraction.txt'
output_file_path = 'C:/Users/100062811/OneDrive - ku.ac.ae/Ghadeer/MatrixMulFullExample/Query_Fraction_hex.txt'
convert_file_to_hex_pure_fraction(input_file_path, output_file_path)
print('Conversion complete.')
```

```python
import numpy as np

def hex_to_signed_int(hex_str, total_bits=16):
    """Convert a signed hexadecimal string to an integer."""
    value = int(hex_str, 16)
    if value >= 2**(total_bits - 1):
        value -= 2**total_bits
    return value

def signed_int_to_hex(value, total_bits=16):
    """Convert a signed integer to a hexadecimal string."""
    if value < 0:
        value = (1 << total_bits) + value
    return format(value & ((1 << total_bits) - 1), '02x')

def multiply_hex_matrices(matrix_a, matrix_b):
    """Multiply two 4x4 matrices with signed hex elements."""
    int_matrix_a = np.array([[hex_to_signed_int(cell) for cell in row] for row in matrix_a])
    int_matrix_b = np.array([[hex_to_signed_int(cell) for cell in row] for row in matrix_b])

    result_matrix = np.dot(int_matrix_a, int_matrix_b)

    hex_result_matrix = [[signed_int_to_hex(cell) for cell in row] for row in result_matrix]
    return hex_result_matrix
    
def add_hex_add_hex_add(int_matrix_a, int_matrix_b,int_matrix_c):
    """Multiply two 4x4 matrices with signed hex elements."""
    
    int_matrix_a = np.array([[hex_to_signed_int(cell) for cell in row] for row in int_matrix_a])
    int_matrix_a= np.left_shift(int_matrix_a, 8)
    # hex_a_matrix = [[signed_int_to_hex(cell) for cell in row] for row in int_matrix_a]
    # print("a after shifting",hex_a_matrix)
    
    int_matrix_b = np.array([[hex_to_signed_int(cell) for cell in row] for row in int_matrix_b])
    
    int_matrix_c = np.array([[hex_to_signed_int(cell) for cell in row] for row in int_matrix_c])
    
    result_matrix = np.add(int_matrix_b, int_matrix_c)
    frac_sum = [[signed_int_to_hex(cell) for cell in row] for row in result_matrix]
    
    result_matrix = np.add(result_matrix, int_matrix_a)

    hex_result_matrix = [[signed_int_to_hex(cell) for cell in row] for row in result_matrix]
    return hex_result_matrix

# Example usage
matrix_a_int = [['02','fd','05','f9'],['04','01','fe','03'],['03','01','09','fb'],['06','00','fc','01']]
matrix_a_frac = [['80','c0','2f','5a'],['73','00','03','19'],['fc','35','a7','20'],['37','90','8e','de']]

matrix_b_int = [['01','f7','06','08'],['07','02','fc','fb'],['fd','05','07','02'], ['04','fa','01','f7']]
matrix_b_frac = [['3b', '71', '21', 'c7'],['1c','c9','5b','c8'],['c6','1c','73','38'], ['74','06','00','00']]

matrix_c_int=[['07','08','fb','05'],['fd','02','02','01'],['06','fa','07','04'],['fc','f7','f7','01']]
matrix_c_frac=[['33','c7','7e','58'],['3b','a6','54','1c'],['8c','03','1c','05'],['dd','5a','4d','03']]

result_matrix_int = multiply_hex_matrices(matrix_a_int, matrix_b_int)
print("\n\nResultant Matrix:")
for row in result_matrix_int:
    print(' '.join(row))
    
result_matrix_frac1= multiply_hex_matrices(matrix_a_int, matrix_b_frac)
print("\n\nResultant Matrix:")
for row in result_matrix_frac1:
    print(' '.join(row))

result_matrix_frac2= multiply_hex_matrices(matrix_a_frac, matrix_b_int)
print("\n\nResultant Matrix:")
for row in result_matrix_frac2:
    print(' '.join(row))


result_add_matrix = add_hex_add_hex_add(result_matrix_int, result_matrix_frac1,result_matrix_frac2)
print("\n\nResultant Matrix:")
for row in result_add_matrix:
    print(' '.join(row))
```
